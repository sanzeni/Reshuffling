# Compare single neuron response in the data and in the model



```python
%reset
#%qtconsole
%load_ext autoreload
%autoreload 2
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib as mpl
import scipy.stats as stats

from scipy.optimize import curve_fit

data_monkeys=np.loadtxt('Monkeys_with_trials.txt');
data_mice=np.loadtxt('Mice_with_trials.txt');


# structure of data set is:
# unit number, contrast, laser, rate, other stuff

DATA=[data_mice,data_monkeys]
species=['Mice','Monkeys']
color=['c','m']
bin_edges=np.arange(-300,300,0.1) 
bin_means=(bin_edges[1::]+bin_edges[0:-1])/2   



model_monkeys=np.load('Simulate_monkeys/RATES_monkeys.npy')
model_mice=np.load('Simulate_mice/RATES_mice_2.npy')
MODEL=[model_mice,model_monkeys]



area=4.0;linethicknes=1.0;
font = {'family' : 'normal','weight' : 'normal','size'   : 7, 'family':'serif','serif':['Arial']}
mpl.rc('font', **font)
mpl.rcParams['lines.linewidth'] =linethicknes
mpl.rcParams['text.usetex'] = False
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['mathtext.rm'] = 'serif'
mpl.rcParams['pdf.fonttype'] = 42

N_stat=10**3;

def phi(x):
    return (0.5*(x+np.abs(x)))
```

# CDF of $r$, CDF of $\Delta r$ , scatter plot $r$ vs $\Delta r$

```python
rows=4
cols=4
fig, axs = plt.subplots(rows,cols, figsize=(1.5*1.875*cols,1.5*1.5*rows), 
                        facecolor='w', edgecolor='k')#,sharex='col')
fig.subplots_adjust(hspace = .5, wspace=.5)
axs = axs.ravel()


idx_laser=-1
for idx_species in [0,1]:
    data=DATA[idx_species]
    model=MODEL[idx_species]
    Con=np.unique(data[:,1])
    Las=np.unique(data[:,2])
    cells_id=np.unique(data[:,0]);
    ##################################################
    ## CDF r
    ##################################################
    xmax=[50,250][idx_species]

    
    Las=[Las[0],Las[-1]]
    Cell_Resp=np.zeros((len(cells_id),len(Con),len(Las),))
    for idx_cell in range(len(cells_id)):
        for idx_con in range(len(Con)):
            for idx_las in range(len(Las)):
                mask=(data[:,0]==cells_id[idx_cell])&(data[:,2]==Las[idx_las])&(data[:,1]==Con[idx_con])
                Trial_Resp=data[mask,3::]
                Cell_Resp[idx_cell,idx_con,idx_las]=np.mean(Trial_Resp[np.isnan(Trial_Resp)==False])
    
    alpha=[0.5,1]
    count=0
    for idx_con in [0,len(Con)-1]:
        hist=np.histogram(Cell_Resp[:,idx_con,0],bin_edges,density=True)
        axs[2*idx_species+(count)*cols].plot(bin_means,np.cumsum(hist[0])*np.diff(bin_edges),c='k')#,label='Con=%d'%Con[idx_con])
        hist=np.histogram(Cell_Resp[:,idx_con,-1],bin_edges,density=True)
        axs[2*idx_species+(count)*cols].plot(bin_means,np.cumsum(hist[0])*np.diff(bin_edges),c=color[idx_species])
        axs[2*idx_species+(count)*cols].axis([0,xmax,0,1])
        axs[2*idx_species+(count)*cols].set_xlabel(r'rate (spk/s)')
        axs[2*idx_species+(count)*cols].set_ylabel('CDF')
        axs[2*idx_species+(count)*cols].text(xmax/3,0.2,'Contrast=%d%%'%Con[idx_con], fontsize=8)

        #idx_r_X,0,idx_cells
        hist=np.histogram(model[idx_con,0,:],bin_edges,density=True)
        axs[2*idx_species+(count)*cols+1].plot(bin_means,np.cumsum(hist[0])*np.diff(bin_edges),c='k')#,label='Con=%d'%Con[idx_con])
        hist=np.histogram(model[idx_con,-1,:],bin_edges,density=True)
        axs[2*idx_species+(count)*cols+1].plot(bin_means,np.cumsum(hist[0])*np.diff(bin_edges),c=color[idx_species])
        axs[2*idx_species+(count)*cols+1].axis([0,xmax,0,1])
        axs[2*idx_species+(count)*cols+1].set_xlabel(r'rate (spk/s)')
        axs[2*idx_species+(count)*cols+1].set_ylabel('CDF')
        axs[2*idx_species+(count)*cols+1].text(xmax/3,0.2,'Contrast=%d%%'%Con[idx_con], fontsize=8)
        
        count=count+1

        
        
    ##################################################
    ## CDF DELTA r
    ##################################################
    
    xmin=[-20,-100][idx_species]
    xmax=[40,200][idx_species]

    
    Las=[Las[0],Las[-1]]
    Cell_Resp=np.zeros((len(cells_id),len(Con),len(Las),))
    for idx_cell in range(len(cells_id)):
        for idx_con in range(len(Con)):
            for idx_las in range(len(Las)):
                mask=(data[:,0]==cells_id[idx_cell])&(data[:,2]==Las[idx_las])&(data[:,1]==Con[idx_con])
                Trial_Resp=data[mask,3::]
                Cell_Resp[idx_cell,idx_con,idx_las]=np.mean(Trial_Resp[np.isnan(Trial_Resp)==False])
    
    count=0
    alpha=[0.3,0.6,1.0]
    for idx_con in [0,2,-1]:
        Delta=Cell_Resp[:,idx_con,-1]-Cell_Resp[:,idx_con,0]
        hist=np.histogram(Delta,bin_edges,density=True)
        axs[2*idx_species+(2)*cols].plot(bin_means,np.cumsum(hist[0])*np.diff(bin_edges),c=color[idx_species],alpha=alpha[count],label='Con=%d'%Con[idx_con])
        
        axs[2*idx_species+(2)*cols].legend(loc='lower right')   #prop={'size': 6},     
        axs[2*idx_species+(2)*cols].axis([xmin,xmax,0,1])
        axs[2*idx_species+(2)*cols].set_xlabel(r'$\Delta$rate (spk/s)')
        axs[2*idx_species+(2)*cols].set_ylabel('CDF')
        axs[2*idx_species+(2)*cols].axvline(x=0,c='k',ls='--')
        
        Delta=model[idx_con,-1,:]-model[idx_con,0,:]
        hist=np.histogram(Delta,bin_edges,density=True)
        axs[2*idx_species+(2)*cols+1].plot(bin_means,np.cumsum(hist[0])*np.diff(bin_edges),c=color[idx_species],alpha=alpha[count],label='Con=%d'%Con[idx_con])
        
        axs[2*idx_species+(2)*cols+1].axis([xmin,xmax,0,1])
        axs[2*idx_species+(2)*cols+1].set_xlabel(r'$\Delta$rate (spk/s)')
        axs[2*idx_species+(2)*cols+1].set_ylabel('CDF')
        axs[2*idx_species+(2)*cols+1].axvline(x=0,c='k',ls='--')

        count=count+1

        
    ##################################################
    ## Scatter
    ##################################################
            
    xmin=[0,0][idx_species]
    xmax=[50,250][idx_species]
    ymin=[-50,-150][idx_species]
    ymax=[50,200][idx_species]
    
    Base=Cell_Resp[:,:,0].ravel()
    Delta=(Cell_Resp[:,:,-1]-Cell_Resp[:,:,0]).ravel()
         
    axs[3*cols+2*idx_species].scatter(Base,Delta,marker='o',c=color[idx_species],s=1,alpha=0.3)
    axs[3*cols+2*idx_species].axhline(y=0,ls='--',c='k')
    axs[3*cols+2*idx_species].set_ylabel('$\Delta$rate (spk/s)')
    axs[3*cols+2*idx_species].set_xlabel('rate without opto (spk/s)')
    axs[3*cols+2*idx_species].axis([xmin,xmax, ymin, ymax])
    axs[3*cols+2*idx_species+1].axis([xmin,xmax, ymin, ymax])
    xxx=np.linspace(0,np.max(Base)/2,10)
    axs[3*cols+2*idx_species].plot(xxx,-xxx,ls='--',c='k')
    
    npercentiles=4
    hist=np.histogram(Base,bin_edges,density=True)
    Base_CDF=np.cumsum(hist[0])*np.diff(bin_edges)

    base_pc=np.zeros(npercentiles)
    delta_pc=np.zeros((npercentiles,2))    
    tests=np.zeros(npercentiles);# Welch's t-test, and F-tesr
    for idx_pc in range(npercentiles):
        lower_bound=bin_means[np.argmin(np.abs(Base_CDF-1./npercentiles*idx_pc))]
        upper_bound=bin_means[np.argmin(np.abs(Base_CDF-1./npercentiles*(idx_pc+1)))]
        mask_base=(Base>=lower_bound)&(Base<upper_bound);
        base_pc[idx_pc]=np.mean(Base[mask_base])
        delta_pc[idx_pc,0]=np.mean(Delta[mask_base])
        delta_pc[idx_pc,1]=stats.sem(Delta[mask_base])
        tests[idx_pc]=(stats.ttest_1samp(Delta[mask_base],0))[1]

    axs[3*cols+2*idx_species].errorbar(base_pc,delta_pc[:,0],delta_pc[:,1],
                                color='k',#color[idx_species],
                                label=species[idx_species])
    
    Base=model[:,0,:].ravel()
    Delta=(model[:,-1,:]-model[:,0,:]).ravel()
         
    axs[3*cols+2*idx_species+1].scatter(Base,Delta,marker='o',c=color[idx_species],s=1,alpha=0.03)
    axs[3*cols+2*idx_species+1].axhline(y=0,ls='--',c='k')
    axs[3*cols+2*idx_species+1].set_ylabel('$\Delta$rate (spk/s)')
    axs[3*cols+2*idx_species+1].set_xlabel('rate without opto (spk/s)')
    
    xxx=np.linspace(0,np.max(Base)/2,10)
    axs[3*cols+2*idx_species+1].plot(xxx,-xxx,ls='--',c='k')
    
    npercentiles=40
    hist=np.histogram(Base,bin_edges,density=True)
    Base_CDF=np.cumsum(hist[0])*np.diff(bin_edges)

    base_pc=np.zeros(npercentiles)
    delta_pc=np.zeros((npercentiles,2))    
    tests=np.zeros(npercentiles);# Welch's t-test, and F-tesr
    for idx_pc in range(npercentiles):
        lower_bound=bin_means[np.argmin(np.abs(Base_CDF-1./npercentiles*idx_pc))]
        upper_bound=bin_means[np.argmin(np.abs(Base_CDF-1./npercentiles*(idx_pc+1)))]
        mask_base=(Base>=lower_bound)&(Base<upper_bound);
        base_pc[idx_pc]=np.mean(Base[mask_base])
        delta_pc[idx_pc,0]=np.mean(Delta[mask_base])
        delta_pc[idx_pc,1]=stats.sem(Delta[mask_base])
        tests[idx_pc]=(stats.ttest_1samp(Delta[mask_base],0))[1]

    axs[3*cols+2*idx_species+1].errorbar(base_pc,delta_pc[:,0],delta_pc[:,1],
                                color='k',#color[idx_species],
                                label=species[idx_species])
    
axs[0].set_title('Mice, data');
axs[1].set_title('Mice, model');
axs[2].set_title('Monkeys, data');
axs[3].set_title('Monkeys, model');
    
    
plt.tight_layout()
fig.set_size_inches(1.5*1.875*cols,1.5*1.5*rows)
pic_name='Stats_data_vs_model.pdf'
fig.savefig(pic_name,format='pdf',bbox_inches='tight')
pic_name='Stats_data_vs_model.png'
fig.savefig(pic_name,format='png',bbox_inches='tight', dpi=300)

```

# Example cells in model

```python
rows=8
cols=7

DeltaT=0.1
Ntrial=[100,20]

idx_laser=-1
for idx_species in [0,1]:
    data=DATA[idx_species]
    model=MODEL[idx_species]
    Con=np.unique(data[:,1])
    Las=np.unique(data[:,2])
    cells_id=np.unique(data[:,0]);
    ##################################################
    ## CDF r
    ##################################################
    xmax=[50,250][idx_species]
    c=['k',color[idx_species]]
    Mod=['N','E','S']
    
    idx_to_use=np.random.randint(0,np.shape(model)[2], len(cells_id))

    fig, axs = plt.subplots(rows,cols, figsize=(1.5*1.875*cols,1.5*1.5*rows), 
                            facecolor='w', edgecolor='k')#,sharex='col')
    fig.subplots_adjust(hspace = .5, wspace=.5)
    axs = axs.ravel()

    Las=[Las[0],Las[-1]]
    
    Cell_Resp=np.zeros((len(cells_id),len(Con),len(Las),2))
    Cell_Mod=np.zeros((len(cells_id),2,))
    for idx_cell in range(len(cells_id)):  
        for idx_las in range(len(Las)):
            for idx_con in range(len(Con)):
                m=model[idx_con,idx_las,idx_to_use[idx_cell]]
                Trial_Resp = np.random.poisson(m*DeltaT, Ntrial[idx_species])/DeltaT
                Cell_Resp[idx_cell,idx_con,idx_las,0]=np.mean(Trial_Resp[np.isnan(Trial_Resp)==False])
                Cell_Resp[idx_cell,idx_con,idx_las,1]=stats.sem(Trial_Resp[np.isnan(Trial_Resp)==False])
            m=Cell_Resp[idx_cell,:,idx_las,0]
            e=Cell_Resp[idx_cell,:,idx_las,1]
            axs[idx_cell].plot(Con, m,c=c[idx_las])
            axs[idx_cell].fill_between(Con,m-e,m+e,alpha=0.3,color=c[idx_las])
        axs[idx_cell].set_xscale('symlog', linthresh=12);
        axs[idx_cell].set_ylim([0,1.2*np.max((Cell_Resp[idx_cell,:,:,0]+Cell_Resp[idx_cell,:,:,1]).ravel())]);
        
        
        for idx_con in [0,-1]:
            m=model[idx_con,0,idx_to_use[idx_cell]]
            Trial_Resp = np.random.poisson(m*DeltaT, Ntrial[idx_species])/DeltaT
            Trial_Resp1=Trial_Resp[np.isnan(Trial_Resp)==False]
            m=model[idx_con,-1,idx_to_use[idx_cell]]
            Trial_Resp = np.random.poisson(m*DeltaT, Ntrial[idx_species])/DeltaT
            Trial_Resp2=Trial_Resp[np.isnan(Trial_Resp)==False]

            s,p=(stats.ttest_ind(Trial_Resp2,Trial_Resp1, equal_var=False))
            if p<0.05:
                Cell_Mod[idx_cell,idx_con]=np.sign(s)
  
        axs[idx_cell].set_title('Low=%s, High=%s'%(Mod[int(Cell_Mod[idx_cell,0])],Mod[int(Cell_Mod[idx_cell,1])])); 
    for idx_cell in np.arange(len(cells_id),rows*cols):
        axs[idx_cell].set_axis_off()
        
        
        
    Cell_Mod=np.zeros((np.shape(model)[2],2,))
    for idx_cell in range(np.shape(model)[2]):  
        
        for idx_con in [0,-1]:
            m=model[idx_con,0,idx_cell]
            Trial_Resp = np.random.poisson(m*DeltaT, Ntrial[idx_species])/DeltaT
            Trial_Resp1=Trial_Resp[np.isnan(Trial_Resp)==False]
            m=model[idx_con,-1,idx_cell]
            Trial_Resp = np.random.poisson(m*DeltaT, Ntrial[idx_species])/DeltaT
            Trial_Resp2=Trial_Resp[np.isnan(Trial_Resp)==False]

            s,p=(stats.ttest_ind(Trial_Resp2,Trial_Resp1, equal_var=False))
            if p<0.05:
                Cell_Mod[idx_cell,idx_con]=np.sign(s)
     
    idx_cell=rows*cols-1
    mask=(Cell_Mod[:,0]==-1)&(Cell_Mod[:,1]==-1)
    axs[idx_cell].text(0.,0.8,'SS cells=%.3f%% (%d/%d)'%(len(Cell_Mod[mask,0])/len(Cell_Mod[:,0]),len(Cell_Mod[mask,0]),len(Cell_Mod[:,0])),fontsize=10)
    mask=(Cell_Mod[:,0]==-1)&(Cell_Mod[:,1]==1)
    axs[idx_cell].text(0.,0.6,'SE cells=%.3f%% (%d/%d)'%(len(Cell_Mod[mask,0])/len(Cell_Mod[:,0]),len(Cell_Mod[mask,0]),len(Cell_Mod[:,0])),fontsize=10)
    mask=(Cell_Mod[:,0]==1)&(Cell_Mod[:,1]==-1)
    axs[idx_cell].text(0.,0.4,'ES cells=%.3f%% (%d/%d)'%(len(Cell_Mod[mask,0])/len(Cell_Mod[:,0]),len(Cell_Mod[mask,0]),len(Cell_Mod[:,0])),fontsize=10)
    mask=(Cell_Mod[:,0]==1)&(Cell_Mod[:,1]==1)
    axs[idx_cell].text(0.,0.2,'EE cells=%.3f%% (%d/%d)'%(len(Cell_Mod[mask,0])/len(Cell_Mod[:,0]),len(Cell_Mod[mask,0]),len(Cell_Mod[:,0])),fontsize=10)
    

    plt.tight_layout()
    fig.set_size_inches(1.5*1.875*cols,1.5*1.5*rows)
    pic_name='Model_cells_%s.pdf'%['Mice','Monkeys'][idx_species]
    fig.savefig(pic_name,format='pdf',bbox_inches='tight')
```

# Check why in the model there are more suppressed and then excited cells
