```python
%reset
#%qtconsole
%load_ext autoreload
%autoreload 2
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib as mpl
import scipy.stats as stats
from scipy.interpolate import interp1d
import os
import functions_optimal as fun

from mpl_toolkits.axes_grid.inset_locator import (inset_axes, InsetPosition,mark_inset)

area=4.0;linethicknes=1.0;

font = {'family' : 'normal','weight' : 'normal','size'   : 7, 'family':'serif','serif':['Arial']}
mpl.rc('font', **font)
mpl.rcParams['lines.linewidth'] =linethicknes
mpl.rcParams['text.usetex'] = False
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['mathtext.rm'] = 'serif'
mpl.rcParams['pdf.fonttype'] = 42

data_mice=np.loadtxt('data/Mice_with_trials.txt');
data_monkeys=np.loadtxt('data/Monkeys_with_trials.txt');

data_both_species=[data_mice,data_monkeys]
# structure of data set is:
# unit number, contrast, laser, rate, other stuff

dataset_both_species,Con_both_species,nCon_both_species,normalization_both_species=[],[],[],[]
for idx_species in range(2):
    dataset,Con,nCon=fun.build_dataset(data_both_species[idx_species])
    normalization=1.0;#np.mean(dataset[0,:,0])
    dataset_both_species=dataset_both_species+[dataset]
    Con_both_species=Con_both_species+[Con]
    nCon_both_species=nCon_both_species+[nCon]
    
    
species=['Mice','Monkeys']
colors=['c','m']

results=np.loadtxt('unstructured/results.txt');

sim_param_all=results[:,0:15]
moments_of_r_sim_all=results[:,15:20]
sim_convergence_all=results[:,20:24]
sim_decay_all=results[:,26]/results[:,27]
sim_species=results[:,28]
sim_idx_J=results[:,29]

sim_curr_E=results[:,30]
sim_curr_I=results[:,31]
sim_curr_tot=results[:,32]
sim_curr_opto=results[:,33]

Possible_Tmax_over_tau_E=np.unique(sim_param_all[:,14])
Possible_sigma_Lambda_over_Lambda=np.unique(sim_param_all[:,10])
Possible_r_X=np.unique(sim_param_all[:,12])
Possible_J=np.unique(sim_param_all[:,11])


results_specific_Js=np.loadtxt('unstructured/results_2.txt');

sim_param_all_Js=results_specific_Js[:,0:15]
moments_of_r_sim_all_Js=results_specific_Js[:,15:20]
sim_convergence_all_Js=results_specific_Js[:,20:24]
sim_decay_all_Js=results_specific_Js[:,26]/results_specific_Js[:,27]
sim_species_Js=results_specific_Js[:,28]
sim_idx_J_Js=results_specific_Js[:,29]

sim_curr_E_Js=results_specific_Js[:,30]
sim_curr_I_Js=results_specific_Js[:,31]
sim_curr_tot_Js=results_specific_Js[:,32]
sim_curr_opto_Js=results_specific_Js[:,33]




results_stc=[np.loadtxt('structured_model/vary_J-animal_mouse.txt'),
            np.loadtxt('structured_model/vary_J-animal_monkey.txt'),
            ];

best_J_unstruct=[0.00013,0.000525]
best_J_struct=[0.001*0.093,0.001*0.526]
```

```python
idx_cnt_to_plot=-1

rows=1;cols=4;
fig, axs = plt.subplots(rows,cols, figsize=(2.5*cols,2.*rows), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.5)
area=4.0;linethicknes=1.0;
axs = axs.ravel()
font = {'family' : 'normal','weight' : 'normal','size'   : 7, 'family':'serif','serif':['Arial']}
mpl.rc('font', **font)
mpl.rcParams['lines.linewidth'] =linethicknes
mpl.rcParams['text.usetex'] = False
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['mathtext.rm'] = 'serif'
mpl.rcParams['pdf.fonttype'] = 42


for idx in range(rows*cols):
    axs[idx].spines['top'].set_visible(False)
    axs[idx].spines['right'].set_visible(False)
    axs[idx].set_xlabel(r'$J$ (mV)')
    axs[idx].set_xlim([0.01,1])
    
axs[0].text(-.2, 1.2,'A', transform=axs[0].transAxes,fontsize=8, fontweight='bold', va='top', ha='right');
axs[1].text(-.2, 1.2,'B', transform=axs[1].transAxes,fontsize=8, fontweight='bold', va='top', ha='right');
axs[2].text(-.2, 1.2,'C', transform=axs[2].transAxes,fontsize=8, fontweight='bold', va='top', ha='right');
axs[3].text(-.2, 1.2,'D', transform=axs[3].transAxes,fontsize=8, fontweight='bold', va='top', ha='right');

for idx_species in [1]:
    dataset=dataset_both_species[idx_species]
    Con=Con_both_species[idx_species]
    nCon=nCon_both_species[idx_species]


    for idx_moment in np.arange(5,7,1):
        if (idx_moment==5):
            panel=axs[0]
            #panel.set_title(['Mice','Monkeys'][idx_species])
            ccc=colors[idx_species]
            ymin=-0.75
            ymax=0.25
            #ymin=-0.75
            #ymax=0.5
            panel.set_ylabel(r'$\rho$')
        if idx_moment<6:
            m,sem=dataset[idx_moment,:,0],dataset[idx_moment,:,1]

        if idx_moment==6:
            panel=axs[1]
            ccc=['k',colors[idx_species]][1]
            ymin=[0,-10][idx_species]
            ymax=[20,20][idx_species]
            #panel.set_title(['Mice','Monkeys'][idx_species])
            panel.set_ylabel('Mean $\Delta$rate\n(spk/s)')
            m,sem=dataset[1,:,0]-dataset[0,:,0],np.sqrt(dataset[1,:,1]**2+dataset[0,:,1]**2) 
                        

        data_y=np.mean(m[idx_cnt_to_plot::])*np.ones(len(Possible_J))
        data_e=np.mean(sem[idx_cnt_to_plot::])*np.ones(len(Possible_J))
        if idx_moment==5:
            panel.plot(Possible_J*10**3,data_y,lw=3,c=ccc,label='data')

        if idx_moment!=5:
            panel.plot(Possible_J*10**3,data_y,
                                               lw=3,c=ccc,
                                               )
        panel.fill_between(Possible_J*10**3,data_y-data_e,data_y+data_e,
                                                 color=ccc,alpha=0.2)

        panel.set_xscale('log')
        panel.axis([0.01,1,ymin,ymax])

    panel.axhline(y=0,ls='--',c='k')
    #panel.set_xlabel(r'$J$ (mV)')

  
indices=np.arange(0,20,1)

idx_species=1
indices_Js=[18,19]

    #indices=np.arange(0,20,1)
dataset=dataset_both_species[idx_species]
Con=Con_both_species[idx_species]
nCon=nCon_both_species[idx_species]

for idx_moment in np.arange(5,7,1):
    clr=colors[idx_species]
    data_y_vs_J=np.zeros(len(Possible_J))

    for idx_J in range(len(Possible_J)):
        mask_cases=(sim_species==idx_species)&(sim_idx_J==idx_J)
        data_x=sim_param_all[mask_cases,12]
        if idx_moment==5:
            data_y=moments_of_r_sim_all[mask_cases,4]/moments_of_r_sim_all[mask_cases,3]**2
            panel=axs[0]

        if idx_moment==6:
            data_y=moments_of_r_sim_all[mask_cases,1]
            panel=axs[1]



        data_y_med=np.zeros((len(np.unique(data_x)),3))
        for idx in range(len(np.unique(data_x))):
            mask_x=data_x==np.unique(data_x)[idx]
            data_y_med[idx,0]=np.mean(data_y[mask_x])
            data_y_med[idx,1]=stats.sem(data_y[mask_x])
        data_y_vs_J[idx_J]=np.mean(data_y_med[idx_cnt_to_plot::,0])


    for idx_J in indices_Js:
        mask_cases=(sim_species_Js==idx_species)&(sim_idx_J_Js==idx_J)
        data_x=sim_param_all_Js[mask_cases,12]
        if idx_moment==5:
            data_y=moments_of_r_sim_all_Js[mask_cases,4]/moments_of_r_sim_all_Js[mask_cases,3]**2

        if idx_moment==6:
            data_y=moments_of_r_sim_all_Js[mask_cases,1]

        data_y_med=np.zeros((len(np.unique(data_x)),3))
        for idx in range(len(np.unique(data_x))):
            mask_x=data_x==np.unique(data_x)[idx]
            data_y_med[idx,0]=np.mean(data_y[mask_x])
            data_y_med[idx,1]=stats.sem(data_y[mask_x])
        data_y_vs_J[idx_J]=np.mean(data_y_med[idx_cnt_to_plot::,0])



    panel.plot(Possible_J[indices]*10**3,data_y_vs_J[indices],c=clr,ls='-')


    if idx_moment==6:
        panel.plot(Possible_J[indices]*10**3,results_stc[idx_species][:,1][indices]-results_stc[idx_species][:,0][indices],c=clr,ls=':')
    if idx_moment!=6:
        panel.plot(Possible_J[indices]*10**3,results_stc[idx_species][:,idx_moment][indices],c=clr,ls=':')


for idx_case in [0,1]:
    data_y_vs_J=np.zeros(len(Possible_J))
    for idx_J in range(len(Possible_J)):
        mask_cases=(sim_species==idx_species)&(sim_idx_J==idx_J)

        data_x=sim_param_all[mask_cases,12]
        if idx_case==0:
            data_y=np.abs(sim_curr_tot/sim_curr_E)[mask_cases,]
            clr='k'
        if idx_case==1:
            data_y=np.abs(sim_curr_opto/sim_curr_E)[mask_cases,]
            clr=['c','m'][idx_species]

        data_y_med=np.zeros((len(np.unique(data_x)),3))
        for idx in range(len(np.unique(data_x))):
            mask_x=data_x==np.unique(data_x)[idx]
            data_y_med[idx,0]=np.mean(data_y[mask_x])
            data_y_med[idx,1]=stats.sem(data_y[mask_x])
        data_y_vs_J[idx_J]=np.mean(data_y_med[idx_cnt_to_plot::,0])


    for idx_J in indices_Js:
        mask_cases=(sim_species_Js==idx_species)&(sim_idx_J_Js==idx_J)

        data_x=sim_param_all_Js[mask_cases,12]
        if idx_case==0:
            data_y=np.abs(sim_curr_tot_Js/sim_curr_E_Js)[mask_cases,]
            clr='k'
        if idx_case==1:
            data_y=np.abs(sim_curr_opto_Js/sim_curr_E_Js)[mask_cases,]
            clr=['c','m'][idx_species]

        data_y_med=np.zeros((len(np.unique(data_x)),3))
        for idx in range(len(np.unique(data_x))):
            mask_x=data_x==np.unique(data_x)[idx]
            data_y_med[idx,0]=np.mean(data_y[mask_x])
            data_y_med[idx,1]=stats.sem(data_y[mask_x])
        data_y_vs_J[idx_J]=np.mean(data_y_med[idx_cnt_to_plot::,0])

    idx_moment=5
    idx_plot=2+idx_case
    if idx_case==0:
        axs[idx_plot].plot(Possible_J[indices]*10**3,data_y_vs_J[indices],c=clr,ls='-') 
        axs[idx_plot].plot(Possible_J[indices]*10**3,results_stc[idx_species][:,idx_moment+1+idx_case][indices],
                           c=clr,ls=':')
    if idx_case==1:
        axs[idx_plot].plot(Possible_J[indices]*10**3,data_y_vs_J[indices],c=clr,ls='-',label='unstructured model') 
        axs[idx_plot].plot(Possible_J[indices]*10**3,results_stc[idx_species][:,idx_moment+1+idx_case][indices],
                           c=clr,ls=':',label='structured model')


    axs[idx_plot].set_xscale('log')
    axs[idx_plot].set_yscale('log', )
    axs[idx_plot].set_xlim([0.01,1])
    axs[idx_plot].set_xlabel(r'$J$ (mV)')


for idx in [2,3,]:
    axs[idx].axis([0.01,1,0.002,2]);

    

      
best_J=[best_J_unstruct,best_J_struct]
for idx_species in [1]:
    for idx in range(4):
        for idx_model in [0,1]:
            axs[idx+0*idx_species].axvline(x=best_J[idx_model][idx_species]*10**3,
                                           ls=['-',':'][idx_model],c=['c','m'][idx_species])
    
    
axs[2].set_ylabel(r'Balance index $\beta$');
axs[3].set_ylabel(r'Optogenetic/excitatory input');
     

plt.tight_layout()
fig.set_size_inches(1.875*cols,1.875*rows)
pic_name='Supp_Figure_6_top.pdf'
fig.savefig(pic_name,format='pdf',bbox_inches='tight', dpi=300)
pic_name='Supp_Figure_6_top.png'
fig.savefig(pic_name,format='png',bbox_inches='tight', dpi=300)

```
