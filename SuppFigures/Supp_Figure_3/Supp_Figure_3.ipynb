```julia
using HDF5
using PyPlot
```

```julia
h5open("suppfig3.h5", "r") do fid
    P = fid["P"]
    global rXPs = read(P,"r_Xs")
    global JPs = read(P,"Js")
    global chPs = Matrix{Bool}(read(P,"chs"))
    global convPs = Matrix{Bool}(read(P,"convs"))
    global μEPs = read(P,"mu_Es")
    global μIPs = read(P,"mu_Is")
    global ΔμEPs = read(P,"delta_mu_Es")
    global ΔμIPs = read(P,"delta_mu_Is")
    global σμE2Ps = read(P,"var_mu_Es")
    global σμI2Ps = read(P,"var_mu_Is")
    global σΔμE2Ps = read(P,"var_delta_mu_Es")
    global σΔμI2Ps = read(P,"var_delta_mu_Is")
    global ρμEΔμEPs = read(P,"cov_mu_E_delta_mu_Es")
    global ρμIΔμIPs = read(P,"cov_mu_I_delta_mu_Is")
    global rEPs = read(P,"r_Es")
    global rIPs = read(P,"r_Is")
    global ΔrEPs = read(P,"delta_r_Es")
    global ΔrIPs = read(P,"delta_r_Is")
    global σrE2Ps = read(P,"var_r_Es")
    global σrI2Ps = read(P,"var_r_Is")
    global σΔrE2Ps = read(P,"var_delta_r_Es")
    global σΔrI2Ps = read(P,"var_delta_r_Is")
    global ρrEΔrEPs = read(P,"cov_r_E_delta_r_Es")
    global ρrIΔrIPs = read(P,"cov_r_I_delta_r_Is")
    
    P2 = fid["P2"]
    global rXP2s = read(P2,"r_Xs")
    global JP2s = read(P2,"Js")
    global chP2s = Matrix{Bool}(read(P2,"chs"))
    global convP2s = Matrix{Bool}(read(P2,"convs"))
    global μEP2s = read(P2,"mu_Es")
    global μIP2s = read(P2,"mu_Is")
    global ΔμEP2s = read(P2,"delta_mu_Es")
    global ΔμIP2s = read(P2,"delta_mu_Is")
    global σμE2P2s = read(P2,"var_mu_Es")
    global σμI2P2s = read(P2,"var_mu_Is")
    global σΔμE2P2s = read(P2,"var_delta_mu_Es")
    global σΔμI2P2s = read(P2,"var_delta_mu_Is")
    global ρμEΔμEP2s = read(P2,"cov_mu_E_delta_mu_Es")
    global ρμIΔμIP2s = read(P2,"cov_mu_I_delta_mu_Is")
    global rEP2s = read(P2,"r_Es")
    global rIP2s = read(P2,"r_Is")
    global ΔrEP2s = read(P2,"delta_r_Es")
    global ΔrIP2s = read(P2,"delta_r_Is")
    global σrE2P2s = read(P2,"var_r_Es")
    global σrI2P2s = read(P2,"var_r_Is")
    global σΔrE2P2s = read(P2,"var_delta_r_Es")
    global σΔrI2P2s = read(P2,"var_delta_r_Is")
    global ρrEΔrEP2s = read(P2,"cov_r_E_delta_r_Es")
    global ρrIΔrIP2s = read(P2,"cov_r_I_delta_r_Is")
end
```

```julia
levels = [-0.45,-0.40,-0.35,-0.30,-0.25,-0.20,-0.15,-0.10,-0.05,-1E-3]
fig,axs = subplots(1,2,figsize=(8,4),dpi=200)

CS = axs[1][:contour](rXPs,JPs,(0.8ρrEΔrEPs+0.2ρrIΔrIPs)./(0.8σΔrE2Ps+0.2σΔrI2Ps),
        levels,
        colors=["0.05","0.15","0.25","0.35","0.45","0.55","0.65","0.75","0.85","0.95"])
clabel(CS,levels,fmt="%.1f",inline_spacing=40,
    manual=[(10^1.6,10^0.0),(10^1.2,10^-0.15),(10^1.3,10^-0.4),(10^1.1,10^-0.5),(10^1.1,10^-1.0)])
axs[1][:contour](rXPs,JPs,chPs,[0.1],colors="black",linestyles="dashed")
axs[1][:contour](rXPs,JPs,convPs,[0.2],colors="black",linestyles="dotted")
axs[1][:set_xscale]("log")
axs[1][:set_yscale]("log")
axs[1][:set_xlabel](L"$r_X$ (spk/s)")
axs[1][:set_ylabel](L"$J$ (mV)")
axs[1][:set_aspect]("equal")

CS = axs[2][:contour](rXP2s,JP2s,(0.8ρrEΔrEP2s+0.2ρrIΔrIP2s)./(0.8σΔrE2P2s+0.2σΔrI2P2s),
        levels,
        colors=["0.05","0.15","0.25","0.35","0.45","0.55","0.65","0.75","0.85","0.95"])
clabel(CS,levels,fmt="%.1f",inline_spacing=40,
    manual=[(10^1.5,10^-0.1),(10^1.2,10^-0.15),(10^1.3,10^-0.5),(10^1.1,10^-0.6),(10^1.1,10^-1.0)])
axs[2][:contour](rXP2s,JP2s,chP2s,[0.1],colors="black",linestyles="dashed")
# axs[2][:contour](rXPs,JPs,convPs,[0.2],colors="black",linestyles="dotted")
axs[2][:set_xscale]("log")
axs[2][:set_yscale]("log")
axs[2][:set_xlabel](L"$r_X$ (spk/s)")
axs[2][:set_ylabel](L"$J$ (mV)")
axs[2][:set_aspect]("equal")

axs[1][:set_title]("K = 500",fontsize=16)
axs[2][:set_title]("K = 2000",fontsize=16)

fig[:tight_layout]()
savefig("Supp_Figure_3.pdf")
```
